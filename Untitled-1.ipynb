```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
opr = input("Enter operator (-, +, /, *): ")
if opr == '+':
    print(num1 + num2)
elif opr == '-':
    print(num1 - num2)    
elif opr == '*':
    print(num1 * num2)    
elif opr == '/':
    print(num1 / num2)    
else:
    print("Invalid operator")
    
```

```python
word = input("Enter a word: ").lower()

vowels = "aeiou"
count = 0

for ch in word:
    if ch in vowels:
        count += 1

print("Number of vowels =", count)
```

```python
word = input("Enter a word: ")

upper_count = 0
lower_count = 0

for ch in word:
    if ch.isupper():
        upper_count += 1
    elif ch.islower():
        lower_count += 1

print("Uppercase letters:", upper_count)
print("Lowercase letters:", lower_count)
```

```python
class Device:
    def __init__(self, brand_name):
        self.brand_name = brand_name

    def device_info(self):
        print("this is a device")


class Phone(Device):
    def __init__(self, brand_name, os):
        super().__init__(brand_name)
        self.os = os

    def device_info(self):
        print("this is a phone")


class SmartPhone(Phone):
    def __init__(self, brand_name, os):
        super().__init__(brand_name, os)

    def device_info(self):
        print("this is a smart phone")


# create object from smart phone
sp = SmartPhone("Samsung", "Android")

# get attributes
print(sp.brand_name)
print(sp.os)

# use device info method
sp.device_info()
```
